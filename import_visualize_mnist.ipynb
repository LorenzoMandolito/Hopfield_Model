```python
import numpy as np

def Input_mnist(P=10):
    import numpy as np
    from keras.datasets import mnist
    # loading
    (X, y), (k,h) = mnist.load_data()
    # Define patterns
    M = np.zeros((10,28*28))
    l    = 0
    flag = 0
    for i in range(0,10):
        for j in range(np.size(y)):
            if y[j] == i:
                if flag == 0:
                    m = j
                    M[l,:] = np.ravel(X[j,:,:]) 
                    flag = 1
                    l += 1
        flag = 0
    for i in range(0,10):
        M[i] /= np.max(M[i])
    mask = M<=0.5
    M[mask] = -1
    M[~mask] = 1
   
    return M


    
M = Input_mnist()
```

```python
def Matrix_mnist(M):
    from matplotlib import pyplot
    M_inv = np.zeros((28,28,10))
    for i in range(0,10):
        M_inv[:,:,i] = M[i].reshape(28,28)
        pyplot.imshow(M_inv[:,:,i], cmap=pyplot.get_cmap('gray'))
        pyplot.show()
    return M_inv

M_inv = Matrix_mnist(M)



hfiwqhefòuwhdcòqwuhc
```
